```python
import pandas as pd

df = pd.read_csv('../../data/churn_train.csv')
df.head()
```

```python
from sklearn.preprocessing import LabelEncoder, OrdinalEncoder, OneHotEncoder, StandardScaler
import sklearn as sl

sl.set_config(display='diagram')

contract = df['Contract'].copy()
fee = df["Monthly_Fee"].copy()

label_encoder = LabelEncoder()
le = label_encoder.fit_transform(contract)
le_df = pd.DataFrame(le)
```

```python
ordinal_encoder = OrdinalEncoder()
oe = ordinal_encoder.fit_transform(df[["Contract", "Internet_Type"]])
oe_df = pd.DataFrame(oe)
oe_df
```

```python
# 1. 인코더 생성 (Sparse 행렬이 아닌 일반 배열로 받기 위해 sparse_output=False 권장)
ohe = OneHotEncoder(sparse_output=False)

# 2. 2차원 데이터로 학습 및 변환 (대괄호 두 개 중요!)
# 6,400행의 데이터를 가진 2차원 배열이 생성됩니다.
ohe_result = ohe.fit_transform(df[['Contract']])

# 3. 컬럼 이름 가져오기 (학습 시 사용한 컬럼명을 리스트로 전달)
# 여기서 ['Contract']를 정확히 넣어줘야 매칭이 됩니다.
column_names = ohe.get_feature_names_out(['Contract'])

# 4. 데이터프레임 생성
ohe_df = pd.DataFrame(ohe_result, columns=column_names)

ohe_df.head()
```
