```python
import pandas as pd

# 데이터를 로드한다.
df = pd.read_csv('../../data/tmdb_5000_movies.csv')
df.head()
```

```python
# 컬럼 width값을 1000으로 주어 안보이는 부분까지 보이게 한다.
pd.set_option('max_colwidth', 1000)
df.head()
```

```python
# 첫번째 열의 장르값이 어떻게 생겼는지 확인한다.
df.iloc[0]['genres']
```

```python
# str을 파이썬 자료구죠 형태로 바꿔주는 함수
from ast import literal_eval

# 데이터가 str 형태라 리스트로 변경해 준다.
result = literal_eval(df.iloc[0]['genres'])
result
```

```python
# 모든 장르 데이터를 리스트 형태로 변환
df['genres'] = df['genres'].apply(literal_eval)
```

```python
# 리스트로 잘 들어갔는지 확인한다.
df.iloc[0]['genres']
```

```python
# 장르의 이름값을 가져와 확인한다.
genres = []
for g in df.iloc[0]['genres']:
    genres.append(g['name'])

genres
```

```python
# 첫번째 열 타이틀 확인 Avatar
df.iloc[0]['title']
```

```python
# 장르를 담을 리스트를 선언한다.
genres = []

# df의 길이 만큼 반복문을 돌린다.
for i in range(len(df)):
    # 장르 이름을 저장할 임시 리스트를 만든다.
    temp = []
    # 반복문을 돌며 장르 이름을 가져온다.
    for g in df.iloc[i]['genres']:
        temp.append(g['name'])
    
    # 리스트에 있는 장르를 공백을 두어 하나의 문자열로 만든다.
    str_data = ' '.join(temp)
    # 장르를 저장한다.    
    genres.append(str_data)

# 데이터 프레임에 업데이트 한다.
df['genres'] = genres
df['genres']
```

```python
# 벡터화 하기 위해 import
from sklearn.feature_extraction.text import CountVectorizer

# min_df 한번이라도 나온 단어면 넣어라 2이면 두개 이상
# ngram_range 한단어에서 두단어까지 만든다.
count_vec = CountVectorizer(min_df = 1, ngram_range=(1, 2))
# 학습 및 적용
genre_mat = count_vec.fit_transform(df['genres'])
```

```python
# 만들어진 단어들 확인
count_vec.vocabulary_
```

```python
# 배열로 변환해 메트릭스 확인
genre_mat.toarray()
```

```python
# 데이터 프레임으로 저장
df_genre_vec = pd.DataFrame(genre_mat.toarray())
# 단어를 정렬해서 컬럼명을 바꾼다
df_genre_vec.columns = sorted(count_vec.vocabulary_, key=count_vec.vocabulary_.get)
# 인덱스는 타이틀값 사용
df_genre_vec.index = df['title']
# 상위 5개 표시
df_genre_vec.head()
```

```python
# 코사인 유사도를 확인하기 위한 함수
from sklearn.metrics.pairwise import cosine_similarity

# 두개의 데이터간 코사인 유사도를 구한다.
c_sim = cosine_similarity(df_genre_vec, df_genre_vec)
```

```python
# 데이터 프레임으로 저장
g_sim_df = pd.DataFrame(c_sim)
# 컬럼명을 타이틀로 한다
g_sim_df.columns = df['title']
# 인덱스를 타이틀로 한다
g_sim_df.index = df['title']
g_sim_df.head()
```

```python
# 슈퍼맨2 의 인덱스값을 구한다.
index = df[df['title'] == 'Superman II'].index.values[0]
# 슈퍼맨2의 비슷한 타이틀을 정렬한다. 1이면 같은 장르 0이면 관련 없음
g_sim_df.iloc[index].sort_values(ascending=False)
```
