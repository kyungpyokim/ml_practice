```python
from sklearn.model_selection import train_test_split
from sklearn.preprocessing import StandardScaler
from sklearn.tree import DecisionTreeClassifier
from sklearn.model_selection import GridSearchCV
from sklearn.preprocessing import LabelEncoder
import numpy as np
```

```python
import pandas as pd

df = pd.read_csv("../../data/titanic.csv")
df_raw = df.copy()
df.head()
```

```python
df["Embarked"].value_counts()
df["Embarked"] = df['Embarked'].fillna(df['Embarked'].mode()[0])
```

```python
df.info()
df.describe()
```

```python
av_age = df["Age"].mean()
print(av_age)
df["Age"] = df["Age"].fillna(av_age)
df.isna().sum()
```

```python
df = df.drop("Cabin", axis=1)
```

```python
df["Age"] = df["Age"].astype("int")
```

```python
df["Embarked"].unique()
```

```python
encoder = LabelEncoder()

df["Sex"] = encoder.fit_transform(df["Sex"])
```

```python
encoder = LabelEncoder()

df["Embarked"] = encoder.fit_transform(df["Embarked"])
```

```python
X_train  = df.loc[:, ["Sex", "Age", "SibSp", "Parch", "Fare", "Embarked"]]
y_train = df["Survived"]
print(X_train, y_train)
```

```python

X_train, X_test, y_train, y_test = train_test_split(X_train, y_train, test_size=0.2, random_state=3333)
print(X_train)
print(X_test)
print(y_train)
print(y_test)
```

```python
model = DecisionTreeClassifier(random_state=3333)

parameters = { "max_depth": [3, 4, 5, 6, 7, 8, 9, 10], "min_samples_split": [2, 3, 4, 5, 10] } # 하이퍼 파라미터

scaler = StandardScaler()
X_train = scaler.fit_transform(X_train)

grid_model = GridSearchCV(model, param_grid=parameters, cv=5)

grid_model.fit(X_train, y_train)
```

```python
grid_model.best_score_
```

```python
grid_model.best_params_
```

```python
best_model = grid_model.best_estimator_
```

```python
X_test = scaler.transform(X_test)

pred = best_model.predict(X_test)
```

```python
from sklearn.metrics import accuracy_score

acc = accuracy_score(pred, y_test)
acc
```
