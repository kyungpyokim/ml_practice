```python
from sklearn.datasets import load_iris
from sklearn.model_selection import cross_val_score
from sklearn.tree import DecisionTreeClassifier
import pandas as pd

iris_data = load_iris()
model = DecisionTreeClassifier(random_state=11)

scores = cross_val_score(model, iris_data.data, iris_data.target, scoring="accuracy", cv=10)
```

```python
scores.mean()
```

```python
from sklearn.datasets import load_iris
from sklearn.tree import DecisionTreeClassifier
from sklearn.model_selection import GridSearchCV, train_test_split

iris_data = load_iris()
X_train, X_test, y_train, y_test = train_test_split(iris_data.data, iris_data.target, stratify=iris_data.target, test_size=0.2)

model = DecisionTreeClassifier(random_state=11)

parameters = { "max_depth": [1, 2, 3, 4, 5], "min_samples_split": [2, 3, 4] } # 하이퍼 파라미터

grid_model = GridSearchCV(model, param_grid=parameters, cv=5)

grid_model.fit(X_train, y_train)
```

```python
grid_model.cv_results_
score_df = pd.DataFrame(grid_model.cv_results_)
score_df
```

```python
grid_model.best_params_
```

```python
grid_model.best_score_
```

```python
grid_model.best_estimator_.predict(X_test)
```
