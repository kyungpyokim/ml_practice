```python
from sklearn.datasets import load_iris
from sklearn.tree import DecisionTreeClassifier # 분류 모델
from sklearn.model_selection import train_test_split

iris_data = load_iris()
iris_data
```

```python
# 몇 개인가? ** 중요 **
# 변수명 : feature -> X lable -> y
X = iris_data.data
X.shape
```

```python
y = iris_data.target
y.shape
```

```python
import pandas as pd

df = pd.DataFrame(data=iris_data.data, columns=iris_data.feature_names)
df.head()
```

```python
df["label"] = iris_data.target
df.head()
```

```python
X_train, X_test, y_train, y_test = train_test_split(iris_data.data, iris_data.target, test_size=0.2, random_state=11) # random_state 같은 데이터셋으로 모델 학습
```

```python
# 갯수 확인 필!!!
print(X_train.shape)
print(y_train.shape)
print(X_test.shape)
print(y_test.shape)
```

```python
# 분류 트리 모델
# ** 같은 random_state 정할것 **
model = DecisionTreeClassifier(random_state=11)
```

```python
model.fit(X_train, y_train)
```

```python
pred = model.predict(X_test)
pred
```

```python
y_test
```

```python
import numpy as np

np.equal(pred, y_test).sum()/30

# 결과는 도메인 마다 다름.
```
