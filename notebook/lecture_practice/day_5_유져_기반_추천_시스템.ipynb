```python
movies_ratings_df = pd.merge(ratings_df, movies_df, on='movieId')
movies_ratings_df.head()
```

```python
user_ratings_df = movies_ratings_df.pivot_table(index='userId', columns='title', values='rating')
user_ratings_df = user_ratings_df.fillna(0)
user_ratings_df.head()
```

```python
from sklearn.metrics.pairwise import cosine_similarity

user_base_reulst = cosine_similarity(user_ratings_df)
user_base_reulst
```

```python
user_base_df = pd.DataFrame(user_base_reulst, index=user_ratings_df.index, columns=user_ratings_df.index)
user_base_df.head()
```

```python
user_base_df = pd.DataFrame(user_base_reulst, index=user_ratings_df.index)
user_base_df.head()
```
