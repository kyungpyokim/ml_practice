```python
genres = []
for i in range(len(df)):
    temp = []
    for g in df.iloc[i]['genres']:
        temp.append(g['name'])
    
    str_data = ' '.join(temp)
    genres.append(str_data)

df['genres'] = genres
df['genres']
```

```python
from sklearn.feature_extraction.text import CountVectorizer

count_vec = CountVectorizer(min_df = 1, ngram_range=(1, 2)) # min_df 한번이라도 나온 단어면 넣어라 2이면 두개 이상
genre_mat = count_vec.fit_transform(df['genres'])
```

```python
count_vec.vocabulary_
```

```python
genre_mat.toarray()
```

```python
df_genre_vec = pd.DataFrame(genre_mat.toarray())
df_genre_vec.columns = sorted(count_vec.vocabulary_, key=count_vec.vocabulary_.get)
df_genre_vec.index = df['title']
df_genre_vec.head()
```

```python
from sklearn.metrics.pairwise import cosine_similarity

c_sim = cosine_similarity(df_genre_vec, df_genre_vec)
```

```python
g_sim_df = pd.DataFrame(c_sim)
g_sim_df.columns = df['title']
g_sim_df.index = df['title']
g_sim_df.head()
```

```python
index = df[df['title'] == 'Superman II'].index.values[0]
g_sim_df.iloc[index].sort_values(ascending=False)
```
