```python
import pandas as pd
import matplotlib.pyplot as plt
```

```python
df = pd.read_csv('../../data/out_visit.csv', index_col='Unnamed: 0')
df.head()
```

```python
plt.figure(figsize=(7, 5))
plt.scatter(df["monthly_spend"], df['visit_count'], alpha=0.2)
plt.xlabel('monthly spend')
plt.ylabel('visit count')
plt.show()
```

```python
# ** DBSCAN은 스케일링 꼭 해줘야 함 **
from sklearn.cluster import DBSCAN
from sklearn.preprocessing import StandardScaler

scaler = StandardScaler()

scaled_data = scaler.fit_transform(df)
scaled_data
```

```python
dbscan_model = DBSCAN(min_samples=8)
labels = dbscan_model.fit_predict(scaled_data)
print('이상치 데이터 갯수 : ', (labels == -1).sum())

labels
```

```python
set(labels)
```

```python
df['cluster'] = labels
df.head()
```

```python
print('군집별 고객수 : ', df['cluster'].value_counts())
```

```python
plt.figure(figsize=(7, 5))
plt.scatter(df["monthly_spend"], df['visit_count'], alpha=0.2, c=df['cluster'])
plt.xlabel('monthly spend')
plt.ylabel('visit count')
plt.show()
```
