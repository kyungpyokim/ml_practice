```python
import folium
import pandas as pd
from folium.plugins import MarkerCluster
```

```python
df = pd.read_csv("../../data/gangnam_starbuks.csv")
df = df.drop("Unnamed: 0", axis=1)
df.head()
```

```python
center_lat = df["Latitude"].mean()
center_lon = df["Longitude"].mean()
```

```python
map = folium.Map(location=[center_lat, center_lon], zoom_start=13)
```

```python
cluster = MarkerCluster().add_to(map)

for _, row in df.iterrows():
    folium.CircleMarker(
        location=[row.Latitude, row.Longitude], radius=3, fill=True, fill_opacity=0.8
    ).add_to(cluster)
```

```python
# map
```

```python
from sklearn.cluster import KMeans

kmeans_model = KMeans(n_clusters=3, random_state=3333)
kmeans_model.fit(df.values)
```

```python
print(f"Expected features: {kmeans_model.n_features_in_}")
```

```python
df.shape
```

```python
kmeans_model.labels_
```

```python
df["cluster"] = kmeans_model.labels_
df.head()
```

```python
df.shape
```

```python
kmeans_model.predict([[37.496495, 127.027475]])
```

```python
kmeans_model.cluster_centers_
```

```python
import matplotlib.pyplot as plt
import seaborn as sns

plt.figure(figsize=(12, 10))

sns.scatterplot(
    x="Longitude",
    y="Latitude",
    hue="cluster",
    data=df,
    palette="viridis",
    s=50,
    alpha=0.6,
    legend="full",
)
plt.scatter(
    kmeans_model.cluster_centers_[:, 1],
    kmeans_model.cluster_centers_[:, 0],
    s=300,
    c="red",
    marker="*",
)

plt.show()
```
