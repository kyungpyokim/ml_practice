```python
import numpy as np
import pandas as pd
from sklearn.preprocessing import StandardScaler
from sklearn.tree import DecisionTreeClassifier
from sklearn.model_selection import GridSearchCV

df_train = pd.read_excel("../data/train_dataset.xlsx")
print(df_train.head())

X_train = df_train.loc[:, ["Age", "Experience(Years)", "Salary(만원)"]]
y_train = df_train["Purchased"]
print(X_train, y_train)

scaler = StandardScaler()
X_train = scaler.fit_transform(X_train)

model = DecisionTreeClassifier(random_state=11)
param_grid = {
    "max_depth": [2, 3, 4, 5],
    "min_samples_split": [2, 3, 4, 5, 6]
}

grid_model = GridSearchCV(model, param_grid=param_grid, cv=5)
grid_model.fit(X_train, y_train)
```

```python
print(grid_model.best_params_)
print(grid_model.best_score_)
```

```python
best_model = grid_model.best_estimator_
```

```python
df_test = pd.read_excel("../data/test_dataset.xlsx")

X_test = df_test.loc[:, ["Age", "Experience(Years)", "Salary(만원)"]]
y_test = df_test["Purchased"]

X_test = scaler.transform(X_test)
pred = best_model.predict(X_test)
pred
```

```python
from sklearn.metrics import accuracy_score

acc = accuracy_score(pred, y_test)
acc
```
